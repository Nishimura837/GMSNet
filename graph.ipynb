```python
import matplotlib.pyplot as plt

# グラフのデータ
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# グラフの作成
plt.plot(x, y, label="y = x^2")

# グラフのラベルとタイトル
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Plot")

# グラフの凡例
plt.legend()

# PDF形式で保存
plt.savefig("mnt/plot.bmp", format="bmp")

# グラフを表示（任意）
plt.show()
```
